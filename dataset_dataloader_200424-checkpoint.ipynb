## Dataset 클래스

```python
#torch.utils.data.Dataset은 데이터셋을 나타내는 추상클래스임
class FaceLandmarksDataset(Dataset): #Dataset상속
    """Face Landmarks dataset"""
    def __init__(self,csv_file,root_dir,transform=None):
        """
        Args:
            csv_file(string):csv파일의 경로
            root_dir(string):모든 이미지가 존재하는 디렉토리 경로
            transform(callable,optional): 샘플에 적용될 Optional transform
        """
        self.landmarks_frame=pd.read_csv(csv_file)
        self.root_dir=root_dir
        self.transform=transform
    #Override    
    def __len__(self):#__len__은 데이터셋의 크기를 리턴해야함
        return len(self.landmarks_frame) 
    
    def __getitem__(self,idx):#dataset[i]에서 호출된
                            #__getitem__은 i번째 샘플을 찾는데 사용됨
        if torch.is_tensor(idx):
            idx=idx.tolist()
        img_name=os.path.join(self.root_dir,self.landmarks_frame.iloc[idx,0])
        image=io.imread(img_name)
        landmarks=self.landmarks_frame.iloc[idx,1:]
        landmarks=np.array([landmarks])
        landmarks=landmarks.astype('float').reshape(-1,2)
        sample={'image':image,'landmarks':landmarks}#데이터셋 샘플은 
                            #{'image':image,'landmarks':landmarks} 의 사전형태임
        
        if self.transform:#필요한 전처리 과정을 샘플에 적용할 수 있음
            sample=self.transform(sample)
            
        return sample
        
```
